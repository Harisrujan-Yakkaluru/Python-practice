```python
# Factorial of a given number
n = int(input())
def factorial(n):
    if n==1 or n==0:
        return 1
    else:
        return n*(factorial(n-1))
factorial(n)
```

```python
# Fibonacci Series
n = int(input())
def fibonacci(n):
    if n==0 or n==1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
fibonacci(n)
```

```python
# Sum of n natural numbers
def sum_of_n(n):
    if n==1:
        return 1
    return n+sum_of_n(n-1)
sum_of_n(5)
```

```python
# Product of n natural numbers
def product_of_n(n):
    if n==1:
        return 1
    return n*product_of_n(n-1)
product_of_n(7)
```

```python
def binary(n):
    if n==0:
        return "0"
    if n==1:
        return "1"
    return binary(n//2)+str(n%2)
binary(30)
```

###### Anonymous function

```python
y = lambda x: x+2
print(y(5))
```

```python
square = lambda x: x**2
print(square(10))
```

```python
# Sum of 2 numbers using lambda
sum = lambda a, b: a + b
print(sum(6, 8))
```

```python
fact = lambda n: n*(n-1)
print(fact(5))
```

```python
# Find square of all list of integers using lambda and map()
arr = [1, 2, 3, 4, 5]
res = list(map(lambda x: x**2, arr))
print(res)
```

```python
tup = (4, 5, 6, 7)
res = tuple(map(lambda y: y**3, tup))
print(res)
```

```python
# Find even numbers in list using lambda and filter()
arr = [1, 2, 3, 4, 5]
res = list(filter(lambda x: x%2==0, arr))
print(res)
```

```python
# Find odd numbers in list using lambda and filter()
arr = [1, 2, 3, 4, 5]
res = list(filter(lambda x: x%2!=0, arr))
print(res)
```

```python
# Single line if-else syntax
# True_exp if condition else False_exp
n = int(input())
res = lambda x: f"{x} - Odd" if x%2 else "Even"
print(res(n))
```
