```python
# Local Scope
# We can access the local scope  variable within the scope, we can't access in global scope
def fun(x):
    x = x+1 # x is local scope
    return x
a = 10
res = fun(a)
print(x)
```

```python
# Enclosing Scope
def outer(x):
    y = 10 # Enclosing Scope
    def inner():
        a = 10 # Local Scope
        z = y+a
        print(f"This is Local Scope: {z}")
    inner()
    y = y+x
    print(f"This is Enclosing Scope: {y}")
n = 5
outer(n)
```

```python
# Global Scope
x = 10 # Global Scope
def outer():
    y = 20 # Enclosing Scope
    def inner():
        z = 30 # Local Scope
        res = z+y+x
        print(f"This is Local Scope: {res}")
    res = y+x
    inner()
    print(f"This is Enclosing Scope: {res}")
outer()
```

```python
# Update the global xcope variable inside the local scope by using 'global' keyword
x = 10 # Global Scope
def outer():
    global x
    x = x+1
    print("This is Local Scope val:", x)
outer()
print("This is Global Scope val:", x)
```

```python
# Trying to update Enclosing Scope variable in Local Scope
x = 10 # Global Scope
def outer():
    y = 20 # Enclosing Scope
    def inner():
        nonlocal y
        z = 30 # Local Scope
        y = y+1
        print(f"This is Local Scope: {y}")
    y = y+x # 30
    inner()
    y = y+x
    print(f"This is Enclosing Scope: {y}")
outer()
```
