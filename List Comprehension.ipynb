```python
# Print 1 to n numbers in a list
n = int(input())
l = []
while(n>0):
    for i in l:
        l.append(i)
print(l)
n -= 1
```

```python
# List Comprehension
# Syntax: [exp for iterator in iterable if condition]
# return 1 to n numbers as list using list comprehension
n = int(input())
print([x for x in range(1, n+1)])
```

```python
n = int(input())
[x for x in range(1, n+1)]
```

```python
# Find the first n even numbers
n = int(input())
res = [i for i in range(1, n+1) if i%2==0]
print(res)
```

```python
n = int(input())
res = [i for i in range(0, 2*n+1) if i%2==0]
print(res)
```

```python
# Return list even or odd, for list of integers
arr = [20, 34, 33, 55, 23, 54]
res = ["Even" if i%2==0 else "Odd" for i in arr]
print(res)
```

```python
# If a number is even then add 1, odd then add 0.5
arr = [20, 34, 33, 55, 23, 54]
res = [i+1 if i%2==0 else i+0.5 for i in arr]
print(res)
```

```python
# Find one dice combinations
print([x for x in range(1, 7)])
```

```python
print([[i, j] for i in range(1, 7) for j in range(1, 7)])
```

```python
# Find the two dice combination if face is 4
print([[i, j] for i in range(1, 7) for j in range(1, 7) if i==4 or j==4])
```

```python
lst = list(map(int, input().split()))
res = ["Positive Even" if i % 2 == 0 and i > 0 else
    "Negative Even" if i % 2 == 0 and i < 0 else
    "Positive Odd" if i % 2 != 0 and i > 0 else
    "Negative Odd" for i in lst]
print(res)
```
