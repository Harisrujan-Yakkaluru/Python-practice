```python
# Types of Arguments
# 1.Positional Arguments: The arguments will assign to parameters as per the position
def sub(x:int, y:int)->int:
    c = x-y
    return c
# Function calling block
a, b = 10, 20
res = sub(a, b)
print(res)
print(sub(b, a))
```

```python
# 2.Keyword Arguments:
def sub(x:int, y:int)->int:
    c = x-y
    return c
# Function calling block
a, b = 10, 20
print(sub(x = a, y = b))
print(sub(y = b, x = a))
```

```python
def addition(a:int, b:int, c:int, d:int)->int:
    s = a+b+c+d
    return s
# Function calling block
w, x, y, z = 2, 3, 4, 5
print(addition(a, b, c, d))
```

```python
# 3.Default Arguments: The parameters defined with some default values, and it assign the argument value, 
# if you pass the values for default parameters otherwise, it will assign default values
def add(a, b, c = 0, d = 0):
    res = a+b+c+d
    return res
print(add(10, 20))
print(add(10, 20, 6))
print(add(10, 20, 2, 1))
```

```python
def multiply(a, b, c = 1, d = 1):
    product = a*b*c*d
    return product
print(multiply(4, 10))
print(multiply(9, 3, 1))
print(multiply(10, 2, 8, 7))
```

```python
# 4.Variable length of positional arguments
def add(a, b = 0, *args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s += num
    return s+a+b
print(add(1, 2, 3, 4, 5, 6, 7, 8, 9))
```

```python
# 5.Variable length of keyword arguments
def details(name, age, *marks, **kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs.keys())
    print(name, age, marks, kwargs)
details("Hari", 21, 95, 92, 97, Gender="Male", Location="Hyderabad")
```

```python
# Pass by value
def fun(x):
    print(f"Before update Address of x : {id(x)}")
    x = x + 1
    print(x)
    print(f"After update the x value, Address of x : {id(x)}")
a = 5
print(f"Before function execution address of a : {id(a)} and value : {a}")
print(fun(a))
print(f"Before function execution address of a : {id(a)} and value : {a}")
```

```python
# Pass by value
def fun(x):
    print(f"Before update Address of x : {id(x)}")
    x.append(100)
    print(x)
    print(f"After update the x value, Address of x : {id(x)}")
a = [1, 2, 3]
print(f"Before function execution address of a : {id(a)} and value : {a}")
print(fun(a))
print(f"Before function execution address of a : {id(a)} and value : {a}")
```
