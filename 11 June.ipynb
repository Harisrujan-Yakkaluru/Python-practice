```python
# Check whether the number is a Prime or not
n = int(input())
count = 0
for i in range(1, n+1):
    if(n%i==0):
        count += 1
if(count==2):
    print(f"{n} is a Prime Number")
else:
    print(f"{n} is not a Prime number")
```

```python
# Check whether the number is a Factorial or not
n = int(input())
for i in range(1, n+1):
    if(n%i==0):
        print(i, end=" ")
```

```python
# Check whether the numbers are odd or even
n1, n2, n3 = map(int, input().split())
if(n1%2==0):
    print("Even")
elif(n2%2==0):
    print("Even")
elif(n3%2==0):
    print("Even")
elif(n1%2!=0):
    print("Odd")
elif(n2%2!=0):
    print("Odd")
elif(n3%2!=0):
    print("Odd")
```

```python
# Using for loop
n1, n2, n3 = map(int, input().split())
for i in n1, n2, n3:
    if(i%2==0):
        print("Even")
    else:
        print("Odd")
```

#### Checking the character

```python
text = input("Enter a Character: ")
vowels = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]
vowels_count = 0
consonants_count = 0
special_characters_count = 0
for character in text:
    if character in vowels:
        vowels_count += 1
    elif character.isalpha():
        consonants_count += 1
    else:
        special_characters_count += 1
print(f"Vowels: {vowels_count}")
print(f"Consonants: {consonants_count}")
print(f"Special Characters: {special_characters_count}")
```

#### Count the number of Even and Odd numbers

```python
number = list(map(int, input("Enter a number: ")))
even_count = 0
odd_count = 0
for i in number:
    if(i%2==0):
        even_count += 1
    else:
        odd_count += 1
print(f"Even Count: {even_count}")
print(f"Odd Count: {odd_count}")
```

```python
number = list(map(int, input("Enter a number: ")))
even_numbers = []
odd_numbers = []
for i in number:
    if(i%2==0):
        even_numbers.append(i)
    else:
        odd_numbers.append(i)
print(f"Even Numbers: {even_numbers}")
print(f"Odd Numbers: {odd_numbers}")
```
