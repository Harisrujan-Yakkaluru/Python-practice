#### Find factors of a number

```python
# Find factors of a number
number = int(input())
factors = []
for i in range(1, number+1):
    if(number%i==0):
        factors.append(i)
if(len(factors)==2):
    print(f'{number} is a Prime Number')
else:
    print(f'{number} is not a Prime Number')
print(factors)
```
