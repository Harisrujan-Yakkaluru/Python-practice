```python
# With return and without parameters
def addition():
    a = 20
    b = 21
    print(a+b)
result = addition()
print(result)
```

```python
# With return and with parameters
def addition(a, b):
    return a+b
result = addition(10, 20)
print(result)
```

```python
# Without return and with parameters
def addition(a, b):
    print(a+b)
result = addition(10, 20)
print(result)
```

```python
# Without return and without parameters
def addition():
    a = 21
    b = 20
    print(a+b)
result = addition()
print(result)
```
