```python
f = {"bread":2, "butter":1}
most_repeated_word = ""
most_repeated_frequency = 0
for word in f:
    if f[word]>most_repeated_frequency:
        most_repeated_word = word
        most_repeated_frequency = f[word]
print(most_repeated_word)
```

```python
# Steps:
# Read the words
# If the word already existed, increment the count, if not assign None
# Add the word to the list
# Print the length of the list to get the unique words
# Print the dictionary to get each and every word frequency
n = int(input())
f = {}
l = []
for i in range(n):
    word = input()
    if word in f:
        f[word] += 1
    else:
        f[word] = 1
        l.append(word)
print(len(l))
for key in f:
    print(f[key],end = " ")
```

#### Built-in Functions:

```python
# The functions definition is provided by python, those functions called as Built-in Functions
```

```python
# abs(): It returns the absolute value
print(abs(10-20))
print(abs(20-10))
```

```python
# round(val, pos): This function rounds the float value to particular positions
print(round(2.2589, 2))
print(round(2.2539, 2))
print(round(2.2539))
print(int(2.2539))
```

```python
# ord(char): It returns ASCII value of character
# chr(ascii_val): It returns the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
# sorted(iterable, reverse=False, key=function): It returns new sorted iterable
# Sort the fruits based on the length of fruit name
fruits = ['Apple', 'Banana', 'Kivi', 'Mango', 'Water Melon']
res = sorted(fruits)
print(res)
print(sorted(fruits, reverse=True))
print(sorted(fruits, reverse=False, key=len)) # It sorts the list based on length of each element
```

```python
# map(function, iterable)
# Find the square of all list of integers using map function
def square(x):
    return x**2
arr = [1, 2, 3, 4, 5]
res = list(map(square, arr))
print(res)
# filter(function, iterable)
```

```python
# filter(function, iterable): It filters the iterable, if the function returns Non_None value, those values will be added to the result
# Find even numbers from list of numbers
def check_even(x):
    if x%2==0:
        return x
lst = [1, 2, 3, 4, 5, 6]
res = list(filter(check_even, lst))
print(res)
```

```python
# Find the 3's factors/multipliers from list of integers using filter funciton
def factors_3(x):
    if x%3==0:
        return x
lst = [1, 24, 54, 22, 44, 51, 33, 15]
res = list(filter(factors_3, lst))
print(res)
```

```python
# enumerate(iterable): This function returns th list of tuple of index and item for iterable elements 
fruits = ['Apple', 'Banana', 'Mango', 'Kivi', 'Watermelon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f"{ind}-{val} length is {len(val)}")
```

```python
# Tuple Unpacking
t = (1, 2)
a, b = t
print(a, b)
```

```python
# zip(iterable1, iterable2, ....): It returns all iterables as single iterable of tuples
roll_no = [101, 102, 103, 104, 105]
names = ['Ravi', 'Raju', 'Siva', 'Rani', 'Gopi']
percentage = [90, 67, 96.8, 67.65, 99.9999]
print(list(zip(roll_no, names, percentage)))
```

```python
# any(): This function returns True when any one iterable items return True
# all(): This function returns True when all iterable items return True
arr = [1, 2, 3, 4, 0]
print(all(arr))
print(any(arr))
```

```python
percentage = [34, 35, 54, 65, 67]
lst = []
for i in percentage:
    if i<35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("All students passed")
else:
    print("All students are not passed")
```
