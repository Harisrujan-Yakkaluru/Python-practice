```python
# Reading integer from user
# Integer to float, string, bool
n = int(input())
print(type(n), n)
print(float(n))
print(bool(n))
print(str(n))
```

```python
o = float(input())
print(type(o), o)
print(int(o))
print(str(o))
print(bool(o))
```

```python
# Reading complex value
real = int(input())
img = int(input())
c = complex(real, img)
print(type(c), c)
```

```python
# Reading string value
string = str(input())
print(type(string), string)
print(int(string))
print(float(string))
print(bool(string))
print(list(string))
print(tuple(string))
print(set(string))
```

```python
s = map(int, input().split())
print(s)
```

```python
# List to tuple, set, string
l = map(int, input().split())
print(l)
print(tuple(l))
print(set(l))
s = str(l)
print(s[0])
```

```python
t = tuple(map(int, input().split()))
print(t)
print(set(t))
```

```python
# Read list of strings
l = list(map(str, input().split()))
print(l)
```
